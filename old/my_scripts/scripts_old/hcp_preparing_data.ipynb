### Dwibiascorrect

```python
%%script tcsh
mrtrixenv
foreach sub (mgh_????)
    foreach shell ('all_shells')
        echo $sub
        echo $shell
        set input = $sub/$shell/data.nii.gz
        set output = $sub/$shell/data_corr.nii.gz
        set bias = $sub/$shell/bias.nii.gz
        set mask = $sub/nodif_brain_mask.nii.gz
        set bvals = $sub/$shell/bvals
        set bvecs = $sub/$shell/bvecs
        echo 'processing' $input 
        dwibiascorrect -mask $mask -bias $bias -ants -fslgrad $bvecs $bvals $input $output
    end
end
```
